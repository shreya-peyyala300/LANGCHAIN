```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI

# Model
model = ChatOpenAI(
    model="openai/gpt-4o-mini",
    base_url="https://openrouter.ai/api/v1",
)

# Prompt to generate movie title
prompt_template = PromptTemplate.from_template(
    "Suggest a {genre} movie in {language}"
)

# Movie title chain (DO NOT invoke yet)
movie_title_chain = prompt_template | model | StrOutputParser()

# Prompt to summarize the movie
movie_summary_prompt = PromptTemplate.from_template(
    "Give me 2-3 line summary of the movie {movie_name}"
)

# Composed chain
from langchain_core.runnables import RunnableLambda

# Debug printer: prints input but passes it unchanged
print_title = RunnableLambda(lambda x: (print("Debug:", x), x)[1])

composed_chain = (
    {"movie_name": movie_title_chain}  # movie_title_chain is a Runnable
    | movie_summary_prompt
    | print_title        # safe debugging
    | model
    | StrOutputParser()
)

response = composed_chain.invoke({
    "language": "Tamil",
    "genre": "Comedy"
})

print("Final summary:", response)
```

```python
from langchain_core.runnables import RunnableSequence
compose_chain=RunnableSequence( {"movie_name": movie_title_chain}, movie_summary_prompt , print_title  , model, StrOutputParser()
)

summary = composed_chain.invoke({"language": "english", "genre":"horror"})
print(summary)
```

```python
from langchain_core.runnables import RunnableParallel

translate_hindi_chain = ChatPromptTemplate.from_template("Translate the summary {summary} to hindi") | model | StrOutputParser()
tranlate_spanish_chain = ChatPromptTemplate.from_template("Translate the summary {summary} to spanish") | model | StrOutputParser()

translate_runnable = RunnableParallel(hindi_translate = translate_hindi_chain, spanish_translate = tranlate_spanish_chain)

translated_summary = translate_runnable.invoke({"summary" : summary})

print("Hindi Summary: ",  translated_summary["hindi_translate"])
print("Spanish Summary: ", translated_summary["spanish_translate"])
```
